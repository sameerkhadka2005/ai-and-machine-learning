## Section 3: MCP Neurons

The McCulloch-Pitts (MCP) neuron, proposed in 1943, is the first computational model of a biological neuron. It is also known as a **linear threshold gate** or **threshold logic unit**.

### Mathematical Formalization:
1. Compute the sum of inputs: $g(X) = \sum_{i=1}^{n} x_i$
2. If any input $x_i$ is **inhibitory**, set output $y = 0$
3. Otherwise: if $g(X) \geq T$, then $y = 1$, else $y = 0$

## Task 1: Implementation of MCP Neurons

### AND Gate using MCP Neuron

````python
def MCP_Neurons_AND(X1, X2, T):
    """
    This function implements basic AND operations with MCP Neuron for two inputs.

    Arguments:
        X1 (1D array): An array of binary values.
        X2 (1D array): An array of binary values.
        T  (int/float): Threshold value.

    Returns:
        state_neuron (1D list): State of neuron (1 or 0) for each input pair.
    """
    assert len(X1) == len(X2), "Input arrays must have the same length."

    # Step 1: Element-wise sum of the two input arrays
    element_sums = [x1 + x2 for x1, x2 in zip(X1, X2)]

    # Step 2: Apply threshold — output 1 if sum >= T, else 0
    state_neuron = []
    for s in element_sums:
        if s >= T:
            state_neuron.append(1)
        else:
            state_neuron.append(0)

    return state_neuron


# Example usage for MCP_Neurons_AND function
X1 = [0, 0, 1, 1]
X2 = [0, 1, 0, 1]
T = 2  # Threshold value for AND gate (both inputs must be 1)

result_and = MCP_Neurons_AND(X1, X2, T)
print(f"Output of AND gate for inputs {X1} and {X2} with threshold {T}: {result_and}")
print()

# Truth table display
print("AND Gate Truth Table:")
print("-" * 30)
print(f"{'X1':>5} {'X2':>5} {'Output':>8}")
print("-" * 30)
for x1, x2, out in zip(X1, X2, result_and):
    print(f"{x1:>5} {x2:>5} {out:>8}")
````

### OR Gate using MCP Neuron

````python
def MCP_Neurons_OR(X1, X2, T):
    """
    This function implements basic OR operations with MCP Neuron for two inputs.

    Arguments:
        X1 (1D array): An array of binary values.
        X2 (1D array): An array of binary values.
        T  (int/float): Threshold value.

    Returns:
        state_neuron (1D list): State of neuron (1 or 0) for each input pair.
    """
    assert len(X1) == len(X2), "Input arrays must have the same length."

    # Step 1: Element-wise sum of the two input arrays
    element_sums = [x1 + x2 for x1, x2 in zip(X1, X2)]

    # Step 2: Apply threshold — output 1 if sum >= T, else 0
    state_neuron = []
    for s in element_sums:
        if s >= T:
            state_neuron.append(1)
        else:
            state_neuron.append(0)

    return state_neuron


# Example usage for MCP_Neurons_OR function
X1 = [0, 0, 1, 1]
X2 = [0, 1, 0, 1]
T = 1  # Threshold value for OR gate (at least one input must be 1)

result_or = MCP_Neurons_OR(X1, X2, T)
print(f"Output of OR gate for inputs {X1} and {X2} with threshold {T}: {result_or}")
print()

# Truth table display
print("OR Gate Truth Table:")
print("-" * 30)
print(f"{'X1':>5} {'X2':>5} {'Output':>8}")
print("-" * 30)
for x1, x2, out in zip(X1, X2, result_or):
    print(f"{x1:>5} {x2:>5} {out:>8}")
````

## Task 1 Questions

### Question 1: Limitations of MCP Neurons

1. **Fixed (hand-crafted) weights and threshold**: The MCP neuron cannot learn from data. Weights and the threshold $T$ must be set manually by the designer — there is no learning algorithm to adjust them automatically.

2. **Binary inputs and outputs only**: MCP neurons only accept binary inputs ($x_i \in \{0, 1\}$) and produce binary outputs. They cannot handle real-valued (continuous) data.

3. **Cannot represent non-linearly separable functions**: The MCP neuron is a linear threshold gate. It cannot solve problems like **XOR** that require a non-linear decision boundary.

4. **No adaptive learning**: Unlike the Perceptron, there is no weight update rule. The model cannot improve its performance based on errors.

5. **Inhibitory inputs have absolute veto power**: Any inhibitory input immediately forces the output to 0, regardless of the other inputs. This is a very rigid and biologically oversimplified assumption.

6. **Equal weighting of inputs**: All excitatory inputs contribute equally to $g(X)$ (weight = 1). There is no mechanism to differentiate the importance of different inputs.

7. **No bias term**: There is no learnable bias, making it harder to shift the decision boundary.

### Question 2: Can XOR be solved with a single MCP Neuron?

**No, XOR cannot be solved by a single MCP neuron.** XOR is not linearly separable — you cannot draw a single straight line to separate the True outputs (0,1) and (1,0) from the False outputs (0,0) and (1,1).

However, XOR **can** be implemented using a **combination of MCP neurons** (a small network):

**XOR decomposition:**  
$XOR(x_1, x_2) = AND(OR(x_1, x_2), \; NAND(x_1, x_2))$

If-else logic to simulate XOR:
```
OR  layer: if (x1 + x2) >= 1 → output_or = 1, else 0
NAND layer: if (x1 + x2) < 2 → output_nand = 1, else 0
AND  layer: if (output_or + output_nand) >= 2 → final = 1, else 0
```

This requires **two layers** of MCP neurons — something a single MCP neuron cannot achieve.

````python
# Bonus: XOR using a network of MCP Neurons
def MCP_Neurons_XOR(X1, X2):
    """
    Implements XOR using a 2-layer MCP Neuron network:
      Layer 1: OR (T=1) and NAND (T=1 applied to NAND logic)
      Layer 2: AND (T=2) of the two Layer 1 outputs
    """
    assert len(X1) == len(X2)
    state_neuron = []
    for x1, x2 in zip(X1, X2):
        # Layer 1
        or_out   = 1 if (x1 + x2) >= 1 else 0   # OR gate
        nand_out = 1 if (x1 + x2) <  2 else 0   # NAND gate
        # Layer 2
        xor_out  = 1 if (or_out + nand_out) >= 2 else 0
        state_neuron.append(xor_out)
    return state_neuron

X1 = [0, 0, 1, 1]
X2 = [0, 1, 0, 1]
result_xor = MCP_Neurons_XOR(X1, X2)
print(f"Output of XOR gate for inputs {X1} and {X2}: {result_xor}")
print()
print("XOR Gate Truth Table (using 2-layer MCP Network):")
print("-" * 30)
print(f"{'X1':>5} {'X2':>5} {'Output':>8}")
print("-" * 30)
for x1, x2, out in zip(X1, X2, result_xor):
    print(f"{x1:>5} {x2:>5} {out:>8}")
````

## Section 4: The Perceptron

The Perceptron was introduced by Frank Rosenblatt in 1958. It extends the MCP neuron by:
- Incorporating **learnable numerical weights**
- An **adaptive learning process** (weight update rule)

It is a **Linear Binary Classifier** that separates data using a linear decision boundary.

**Mathematical formulation:**
$$z = \sum_{i=1}^{n} w_i x_i + b$$

**Activation (step) function:**
$$\hat{y} = \begin{cases} 1 & \text{if } z \geq 0 \\ 0 & \text{if } z < 0 \end{cases}$$

**Weight update rule (Perceptron Learning Rule):**
$$\Delta w_j = \eta \times (y - \hat{y}) \times x_j$$
$$\Delta b = \eta \times (y - \hat{y})$$

## Task 2: Perceptron Algorithm for 0 vs 1 Classification (MNIST)

### Step 1: Import Libraries

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("Libraries imported successfully.")
````

### Step 2: Load the Dataset

````python
# Load the dataset
# Update the file path if necessary
df_0_1 = pd.read_csv("/content/mnist_0_and_1.csv")

# Extract features and labels
X = df_0_1.drop(columns=["label"]).values  # 784 pixels (28x28 image flattened)
y = df_0_1["label"].values                 # Labels (0 or 1)

# Check the shape of the features and labels
print("Feature matrix shape:", X.shape)
print("Label vector shape:  ", y.shape)
````

#### Answer to Questions 1 & 2:

**Question 1: What does the shape of X represent?**  
`X.shape` is `(n_samples, 784)`. The first dimension is the **number of images** (data points) in the dataset. The second dimension (784) is the **number of features per image** — each 28×28 pixel image is flattened into a vector of 784 pixel intensity values.

**Question 2: What does the shape of y represent?**  
`y.shape` is `(n_samples,)`. It is a 1D array of **class labels** corresponding to each image — either `0` or `1` for the respective digit.

### Step 3: Visualize the Dataset

````python
# Separate images for label 0 and label 1
images_0 = X[y == 0]  # All images with label 0
images_1 = X[y == 1]  # All images with label 1

fig, axes = plt.subplots(2, 5, figsize=(10, 5))

if len(images_0) < 5 or len(images_1) < 5:
    print("Error: Not enough images in images_0 or images_1 to plot 5 images.")
else:
    for i in range(5):
        # Plot digit 0
        axes[0, i].imshow(images_0[i].reshape(28, 28), cmap="gray")
        axes[0, i].set_title("Label: 0")
        axes[0, i].axis("off")

        # Plot digit 1
        axes[1, i].imshow(images_1[i].reshape(28, 28), cmap="gray")
        axes[1, i].set_title("Label: 1")
        axes[1, i].axis("off")

plt.suptitle("First 5 Images of 0 and 1 from MNIST Subset", fontsize=13)
plt.tight_layout()
plt.show()
````

### Step 4: Initialize Weights and Bias

````python
# Initialize weights and bias
weights       = np.zeros(X.shape[1])  # 784 weights (one for each pixel)
bias          = 0
learning_rate = 0.1
epochs        = 100

print(f"Weights shape   : {weights.shape}")
print(f"Initial bias    : {bias}")
print(f"Learning rate   : {learning_rate}")
print(f"Epochs          : {epochs}")
````

#### Answer to Questions 3 & 4:

**Question 3: What does the weights array represent in this context?**  
The `weights` array has one weight per pixel (784 total). Each weight $w_i$ represents the **importance or contribution** of pixel $i$ toward the classification decision. During training, weights associated with pixels that are more discriminative between digits 0 and 1 will grow larger in magnitude.

**Question 4: Why are we initializing the weights to zero? What effect could this have on the training process?**  
Initializing weights to zero is a simple and common starting point. For a single-layer perceptron, it works reasonably well because:
- The perceptron learning rule updates weights **only when a misclassification occurs**, so asymmetry emerges naturally.
- Zero initialization means all weights start equal, which is acceptable since the step function still allows updates when a mistake is made.

However, for deeper networks, zero initialization can cause the **symmetry problem** — all neurons learn the same features and never differentiate. For perceptrons, the main side effect is that the first prediction will always be 0 (since $z = 0$ gives $\hat{y} = 1$ under $z \geq 0$), which may cause more updates in early epochs, but training will still converge.

### Step 5: Implement the Decision (Activation) Function

````python
def decision_function(X, weights, bias):
    """
    Compute predicted labels for input data using a step activation function.

    Parameters:
        X       : numpy array of shape (n_samples, n_features)
        weights : numpy array of shape (n_features,)
        bias    : scalar bias term

    Returns:
        y_pred_all : numpy array of predicted labels (0 or 1)
    """
    # Compute the weighted sum (net input) for all samples at once
    predictions = np.dot(X, weights) + bias

    # Apply step activation function: 1 if net input >= 0, else 0
    y_pred_all = np.where(predictions >= 0, 1, 0)

    return y_pred_all

print("decision_function defined successfully.")
````

### Step 6: Implement the Perceptron Learning Algorithm

````python
def train_perceptron(X, y, weights, bias, learning_rate=0.1, epochs=100):
    """
    Train the perceptron using the Perceptron Learning Algorithm.

    Parameters:
        X             : Features (n_samples, n_features)
        y             : True labels (n_samples,)
        weights       : Initial weights (n_features,)
        bias          : Initial bias (scalar)
        learning_rate : Step size for weight updates (default 0.1)
        epochs        : Number of training iterations (default 100)

    Returns:
        weights  : Updated weights after training
        bias     : Updated bias after training
        accuracy : Final training accuracy (proportion correctly classified)
    """
    n_samples = X.shape[0]

    for epoch in range(epochs):
        convergence_flag = True  # Assume converged unless a mistake occurs

        for i in range(n_samples):
            # Step 1: Compute net weighted input
            output = np.dot(X[i], weights) + bias

            # Step 2: Apply step activation function
            y_hat = 1 if output >= 0 else 0

            # Step 3: Update weights and bias if prediction is wrong
            if y_hat != y[i]:
                error = y[i] - y_hat
                weights = weights + learning_rate * error * X[i]   # Weight update
                bias    = bias    + learning_rate * error           # Bias update
                convergence_flag = False

        # Early stopping if no updates were made in this epoch
        if convergence_flag:
            print(f"Converged at epoch {epoch + 1}")
            break

    # Compute final training accuracy
    y_pred_final = decision_function(X, weights, bias)
    accuracy = np.mean(y_pred_final == y)

    return weights, bias, accuracy


print("train_perceptron function defined successfully.")
````

#### Answer to Questions 5, 6 & 7:

**Question 5: What is the purpose of `output = np.dot(X[i], weights) + bias`?**  
This line computes the **net input** (weighted sum) for sample $i$. It multiplies each feature value $x_j$ by its corresponding weight $w_j$, sums them all up, and adds the bias term. The result $z = \sum w_j x_j + b$ is then passed to the activation function to produce a prediction. Essentially, it measures how strongly the current weights support classifying the sample as class 1.

**Question 6: What happens when the prediction is wrong? How are weights and bias updated?**  
When $\hat{y} \neq y$ (a misclassification), the error $e = y - \hat{y}$ is computed. The update rule is:
- $w_j \leftarrow w_j + \eta \cdot e \cdot x_j$ — each weight shifts toward correcting the error, scaled by the input
- $b \leftarrow b + \eta \cdot e$ — the bias shifts in the direction that reduces the error

If we predicted 0 but the true label is 1 ($e = +1$), weights are increased, making future predictions more likely to be 1. If we predicted 1 but should be 0 ($e = -1$), weights decrease.

**Question 7: Why is the final accuracy important, and what do you expect it to be?**  
Final accuracy tells us how well the perceptron has learned to separate the two classes on the **training data**. Since 0 and 1 are visually very distinct digits, the data is nearly linearly separable, so we expect the perceptron to achieve **very high accuracy — close to 99–100%** after sufficient training epochs. This validates that the perceptron learning algorithm converged successfully.

### Step 7: Train the Perceptron

````python
# Re-initialize weights and bias before training
weights       = np.zeros(X.shape[1])
bias          = 0
learning_rate = 0.1
epochs        = 100

# Train the perceptron
weights, bias, accuracy = train_perceptron(X, y, weights, bias,
                                           learning_rate=learning_rate,
                                           epochs=epochs)

print(f"The Final Training Accuracy is: {accuracy:.4f} ({accuracy*100:.2f}%)")
````

### Step 8: Visualize Misclassified Images

````python
# Get predictions for all data points
predictions = np.dot(X, weights) + bias
y_pred      = np.where(predictions >= 0, 1, 0)

# Calculate final accuracy
final_accuracy = np.mean(y_pred == y)
print(f"Final Accuracy: {final_accuracy:.4f}")

# Visualize Misclassified Images
misclassified_idx = np.where(y_pred != y)[0]

if len(misclassified_idx) > 0:
    print(f"Number of misclassified images: {len(misclassified_idx)}")
    fig, axes = plt.subplots(2, 5, figsize=(10, 5))
    for ax, idx in zip(axes.flat, misclassified_idx[:10]):
        ax.imshow(X[idx].reshape(28, 28), cmap="gray")
        ax.set_title(f"Pred: {y_pred[idx]}, True: {y[idx]}")
        ax.axis("off")
    plt.suptitle("Misclassified Images (0 vs 1)", fontsize=13)
    plt.tight_layout()
    plt.show()
else:
    print("All images were correctly classified!")
````

#### Answer to Questions 8 & 9:

**Question 8: What does `misclassified_idx` store, and how is it used in this code?**  
`misclassified_idx` is a numpy array containing the **indices** of all samples where the perceptron's prediction (`y_pred[idx]`) did not match the true label (`y[idx]`). It is computed using `np.where(y_pred != y)[0]`. It is then used to:
- Count total misclassifications
- Index back into `X` to retrieve the corresponding images
- Display up to 10 of those images with their predicted and true labels as titles

**Question 9: How do you interpret `"All images were correctly classified!"`?**  
This message means the perceptron found a **perfect linear decision boundary** that separates all 0s from all 1s in the training dataset. Since 0 and 1 are the most visually distinct MNIST digits, the data is highly linearly separable, and the perceptron is expected to converge to zero training error. Note that this says nothing about **generalization** to unseen test data — perfect training accuracy does not guarantee perfect test accuracy.

## Task 3: Perceptron Algorithm for 3 vs 5 Classification (MNIST)

````python
# Train the perceptron on the 3 vs 5 dataset
weights_35, bias_35, accuracy_35 = train_perceptron(
    X_35, y_35_binary,
    weights_35, bias_35,
    learning_rate=learning_rate_35,
    epochs=epochs_35
)

print(f"Final Training Accuracy (3 vs 5): {accuracy_35:.4f} ({accuracy_35*100:.2f}%)")
````

### Step 5: Visualize Misclassified Images

````python
# Get predictions for all data points
predictions_35 = np.dot(X_35, weights_35) + bias_35
y_pred_35      = np.where(predictions_35 >= 0, 1, 0)

# Map binary predictions back to original labels for display
y_pred_labels  = np.where(y_pred_35 == 1, 5, 3)

# Final accuracy
final_accuracy_35 = np.mean(y_pred_35 == y_35_binary)
print(f"Final Accuracy: {final_accuracy_35:.4f}")

# Visualize Misclassified Images
misclassified_idx_35 = np.where(y_pred_35 != y_35_binary)[0]

if len(misclassified_idx_35) > 0:
    print(f"Number of misclassified images: {len(misclassified_idx_35)}")
    fig, axes = plt.subplots(2, 5, figsize=(10, 5))
    for ax, idx in zip(axes.flat, misclassified_idx_35[:10]):
        ax.imshow(X_35[idx].reshape(28, 28), cmap="gray")
        ax.set_title(f"Pred: {y_pred_labels[idx]}, True: {y_35[idx]}")
        ax.axis("off")
    plt.suptitle("Misclassified Images (3 vs 5)", fontsize=13)
    plt.tight_layout()
    plt.show()
else:
    print("All images were correctly classified!")
````

### Conclusion: 3 vs 5 Classification

**Observations:**

1. **Harder task than 0 vs 1**: The digits 3 and 5 share many visual similarities — both are curved, closed on the right, and have similar pixel distributions. This makes them harder to separate with a linear boundary.

2. **Lower accuracy expected**: Unlike 0 vs 1, the perceptron will likely **not achieve 100% accuracy** on 3 vs 5. The two classes are not perfectly linearly separable, meaning the perceptron may not converge and will still misclassify some ambiguous-looking digits.

3. **Misclassified images**: The misclassified samples tend to be the most ambiguous ones — a poorly written '3' might look like a '5' and vice versa. The perceptron boundary cannot adapt to the curvature of these differences.

4. **Limitation of the Perceptron**: This demonstrates the key limitation — the perceptron can only find **linear decision boundaries**. For more visually similar or non-linearly separable digit pairs, more powerful models (e.g., multi-layer neural networks) are required.
